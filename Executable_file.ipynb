```python
import heap_sort as hs
import pandas as pd

# This assumes 'patient_df' exists from running the previous cell.

# 1. Create an empty list to use as our priority queue (min-heap)
patient_priority_queue = []

# Check if patient_df exists before using it
if 'patient_df' not in locals() and 'patient_df' not in globals():
    print("Error: 'patient_df' not found.")
    print("Please run the data generation cell (generate_dataset.py) first.")
else:
    print(f"Loading {len(patient_df)} patients into the priority queue...")

    # 2. Iterate through the DataFrame and add patients to the heap
    for row in patient_df.itertuples():
        
        # --- Column Access Guide (based on generate_dataset.py) ---
        # DataFrame Order:
        # Col 1: "patient ID"       -> row._1
        # Col 2: "patient Name"       -> row._2
        # Col 3: "Priority"          -> row.Priority (valid identifier)
        # Col 4: "Department"        -> row.Department (valid identifier)
        # Col 5: "Date of Admission" -> row._5
        # Col 6: "Days Admitted"     -> row._6
        # Col 7: "Current Status"    -> row._7
        
        # 3. Create a tuple for each patient.
        # The heap will sort by the FIRST element, then the SECOND, etc.
        # This order is crucial for correct prioritization.
        patient_tuple = (
            row.Priority,       # 0: Sort by Priority (col 3)
            row._5, # 1: Then by Date of Admission (col 5)
            row._1, # 2: Then by patient ID (col 1)
            
            # --- The rest is just data we carry along ---
            row._2, # 3: patient Name (col 2)
            row.Department, # 4: Department (col 4)
            row._6, # 5: Days Admitted (col 6)
            row._7  # 6: Current Status (col 7)
        )
        
        # 4. Push the patient tuple onto the heap using OUR function
        hs.heap_push(patient_priority_queue, patient_tuple)

    print("Priority queue created successfully.")
    print(f"Total patients in queue: {len(patient_priority_queue)}")

    # --- Let's test it! ---
print("\n--- Processing the Top 10 Highest-Priority patients ---")

num_to_process = min(10, len(patient_priority_queue))

# 1. Create an empty list to store the patients we pop
processed_patients_list = [] 

if num_to_process == 0:
    print("Queue is empty, no patients to process.")
else:
    for i in range(num_to_process):
        # 2. Pop the patient from the heap
        # The popped_patient is the tuple you defined earlier
        popped_patient = hs.heap_pop(patient_priority_queue)
        
        # 3. Add the popped patient tuple to our new list
        processed_patients_list.append(popped_patient)

# 4. Now, AFTER the loop, create a DataFrame from the list of tuples
if processed_patients_list: # Check if the list is not empty
    
    # Define the column names in the *exact same order* as your tuple
    column_names = [
        "Priority",
        "Date of Admission",
        "Patient ID",
        "Patient Name",
        "Department",
        "Days Admitted",
        "Current Status"
    ]
    
    # Create the DataFrame
    processed_df = pd.DataFrame(
        processed_patients_list, 
        columns=column_names
    )
    
    # 5. Print the new DataFrame
    # Using .to_string() ensures it prints nicely without truncation
    print("\n--- Processed Patients DataFrame ---")
    print(processed_df.to_markdown())

print(f"\nRemaining patients in queue: {len(patient_priority_queue)}")

```
